```python
#Target distribution
sns.countplot(x="Attrition",data=df)
plt.title("Attrition Distribution")
plt.show()

print(df["Attrition"].value_counts(normalize=True))
```

This 4:1 ratio represents a moderate to severe class imbalance that will significantly impact our machine learning model performance. The minority class (employees who leave) represents only about 16% of your dataset, which is typical for employee attrition

### Feature Exploration

```python
sns.countplot(x="OverTime",hue="Attrition",data=df)
plt.title("Attrition vs Overtime")
plt.show()
```

As we can see Employees who are not working overtime tends to stay with the company (higher count in orange for "No" attrtion )

```python
sns.boxplot(x="Attrition", y="MonthlyIncome", data=df)
plt.title("Attrition vs MonthlyIncome")
plt.show()
```

The plot shows that employees who left ("Yes") generally have a lower median monthly income and lower overall income range than employees who stayed ("No"), whose income distribution is shifted higher and has more high-value outliers

```python
sns.histplot(df[df["Attrition"]=="Yes"]["Age"], bins=20, color="red", alpha=0.6, label="Attrition=Yes")
sns.histplot(df[df["Attrition"]=="No"]["Age"], bins=20, color="blue", alpha=0.6, label="Attrition=No")
plt.legend()
plt.title("Age Distribution by Attrition")
plt.show()
```

This shows younger employees <30 years of age have higher tendency to leave (more noticeable red bars at lower ages).

```python
#correlation heatmap for humerical feature 
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap (Numeric Features)")
plt.show()
```

The heatmap reveals that time-based features (TotalWorkingYears, YearsAtCompany, YearsInCurrentRole, YearsSinceLastPromotion) and career progression metrics (JobLevel, MonthlyIncome) are highly inter-related and form distinct clusters
